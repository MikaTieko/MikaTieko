```python
import pandas as pd
```

```python
df_principal = pd.read_excel("/content/Tabela de ações-.xlsx", sheet_name="Principal")
df_principal

```

```python
df_total_acoes = pd.read_excel("/content/Tabela de ações-.xlsx", sheet_name="Total_de_acoes")
df_total_acoes
```

```python
df_ticker = pd.read_excel("/content/Tabela de ações-.xlsx", sheet_name="Ticker")
df_ticker
```

```python
df_Chatgpt = pd.read_excel("/content/Tabela de ações-.xlsx", sheet_name="Chatgpt")
df_Chatgpt
```

```python
df_Chatgpt=df_Chatgpt[["Nome da empresa","Segmento","Idade (em anos)"]].copy()
df_Chatgpt
```

```python
df_principal=df_principal[["Ativo","Data","Último (R$)","Var. Dia (%)"]].copy()
df_principal
```

```python
df_principal=df_principal.rename(columns={"Último (R$)":"valor_final","Var. Dia (%)":"var_dias_pct"}).copy()
df_principal
```

```python
df_principal["Var_pct"]=df_principal["var_dias_pct"]/100
df_principal["Valor_Inicial"]=df_principal['valor_final']/(df_principal["Var_pct"]+1)
df_principal
```

```python
df_principal=df_principal.merge(df_total_acoes,left_on="Ativo",right_on="Código",how="left")
df_principal
```

```python
df_principal=df_principal.drop(columns=["Código"])
df_principal
```

```python
df_principal["Variacao_rs"]=(df_principal["valor_final"]-df_principal["Valor_Inicial"])*df_principal["Qtde. Teórica"]
df_principal
```

```python
pd.options.display.float_format="{:.2f}".format
```

```python
df_principal["Qtde. Teórica"]=df_principal["Qtde. Teórica"].astype(int)
df_principal
```

```python
df_principal=df_principal.rename(columns={"Qtde. Teórica":"Qtde_teorica"}).copy()
df_principal
```

```python
df_principal['Resultado'] = df_principal['Var_pct'].apply(lambda x: 'subiu' if x > 0 else ('desceu' if x < 0 else 'estavel'))
df_principal
```

```python
df_principal=df_principal.merge(df_ticker,left_on="Ativo",right_on="Ticker",how="left")
df_principal=df_principal.drop(columns=["Ticker"])
df_principal
```

```python
df_Chatgpt
```

```python
df_principal=df_principal.merge(df_Chatgpt,left_on="Nome",right_on="Nome da empresa",how="left")
df_principal=df_principal.drop(columns=["Nome da empresa"])
df_principal
```

```python
df_principal=df_principal.rename(columns={"Idade (em anos)":"idade"}).copy()
df_principal
```

```python
df_principal['Cat_Idade'] = df_principal['idade'].apply(lambda x: 'mais de 100 anos' if x > 100 else ('menos de 50 anos' if x < 50 else 'entre 50 e 100 anos'))
df_principal
```

```python
maior_variacao = df_principal['Variacao_rs'].max()
menor_variacao = df_principal['Variacao_rs'].min()
media_variacao = df_principal['Variacao_rs'].mean()
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()
print(f'Maior valor de variação: R$ {maior_variacao:,.2f}')
print(f'Menor valor de variação: R$ {menor_variacao:,.2f}')
print(f'Média: R$ {media_variacao:,.2f}')
print(f'Média de quem Subiu: R$ {media_subiu:,.2f}')
print(f'Média de quem Desceu: R$ {media_desceu:,.2f}')
```
